```python
#Q1-List all student info. If student is faculty,list faculty information aswell
#Hint Use left Join
txt='''
Select *
From Student left join Faculty on Student.StdNo=Faculty.FacNo
'''
Q1 =  psql.sqldf(txt)
Q1
```

```python
#Q2-List all Faculty_1 info. If faculty has supervisor ,display related supervisor information from faculty table
#Hint Use left Join
txt='''
select *
from Faculty_1 left join Faculty on Faculty_1.FacSupervisor=Faculty.FacNo

'''
Q2 =  psql.sqldf(txt)
Q2
```

```python
#Q3-List all courses who are not offered 
#Hint Use left Join and look for one of column in Course Table whose value is null.
txt='''
Select *
From Course left join Offering on Course.CourseNo=Offering.CourseNo
where Offering.CourseNo is null

'''
Q3 =  psql.sqldf(txt)
Q3
```

```python
#Q4-List all offerings and the faculty assigned to teach them .Also include courses without a faculty assigned to them and Course Name as well 
#Hint:Use Join for 3 tables. Search internet how to do that
#I have to use left join in order to connect Offering and Faculty Table
#I use inner join in order to connect Course and Offering because I 
#would like to list only offered courses with and without assigned faculty.
txt='''
Select *
From Course inner join Offering on Course.CourseNo=Offering.CourseNo 
 left join Faculty on Offering.FacNo=Faculty.FacNo
'''
Q4 =  psql.sqldf(txt)
Q4
```

```python
#Q5-List courses offfered but not taken any students. 
txt='''
Select *
from Course left join Offering on Course.Courseno=Offering.Courseno
where Offering.Courseno is null
'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#Q6-List  Student LastName and FirstName of the Student with StdGPA greater than average GPA
#Use nested query by calculating average GPA first.
txt='''
Select StdFirstName,	StdLastName	
From Student
where StdGPA >(select avg(StdGPA) from Student )

'''
Q6 =  psql.sqldf(txt)
Q6
```

```python
#Q7a-List Faculty information with a maximium and minimum salary
#Hint use nested query to calculate max and min in the form like- where FacSalary=(...) or FacSalary=(....)
txt='''
Select *
From Faculty
where FacSalary=(select min(FacSalary) from Faculty) or FacSalary=(select max(FacSalary) from Faculty)
'''
Q7a =  psql.sqldf(txt)
Q7a
```

```python
#Q7b-List Faculty information having salary greater than average salary  and minimum salary
#Hint use nested query to calculate max and min in the form like- where FacSalary=(...) or FacSalary=(....)
txt='''
select *
From Faculty
where FacSalary>(select avg(FacSalary) from Faculty) or FacSalary=(select min(FacSalary) from Faculty)

'''
Q7b =  psql.sqldf(txt)
Q7b
```

```python
#Q8- List Faculty information of Faculty who are Supervisor of other faculty.
#This is the second column of Faculty_1 Table.For this reason, we will use
# nested query by finding FacSupervisor where FacSupervisor is not null first
txt='''
select *
from Faculty
where facno  in (select FacSupervisor  from Faculty_1 where FacSupervisor  is not null )
'''
Q8 =  psql.sqldf(txt)
Q8
```

```python
#Q9- List Faculty information of Faculty who are NOT supervisor of other faculty.
#This is similar to Q8 but we want Faculty who are NOT supervisor of other faculty.
#For this reason,we find FacSupervisor No at first in the nested query
#and find faculty information if Facsupervisor are NOT included in FacNo . 
txt='''
select *
from Faculty
where facno not in (select FacSupervisor  from Faculty_1 where FacSupervisor  is not null )

'''
Q9 =  psql.sqldf(txt)
Q9
```

```python
#Q10- List Faculty information of Faculty who has  Supervisor 
#Different than Q8, we will find  FacNo when FacSupervisor is not null 
#as  these are Faculty who has  Supervisor 
txt='''
select *
from Faculty
where facno in (select facno from Faculty_1 where FacSupervisor is not null)
'''
Q10 =  psql.sqldf(txt)
Q10
```

```python
#Q11- List Faculty information of Faculty who has not Supervisor 
#Simmilar to Q10 but we will find  FacNo when FacSupervisor is  null 
txt='''
select *
from Faculty
where facno  in (select facno from Faculty_1 where FacSupervisor is null)

'''
Q11 =  psql.sqldf(txt)
Q11
```

```python
#Q12-Get number of faculty and average salary for each department. 
txt='''
Select FacDept,avg(FacSalary)
from Faculty 
Group by FacDept

'''
Q12 =  psql.sqldf(txt)
Q12
```

```python
#Q13-Get number of faculty, number of course offered  for each faculty rank.
txt='''
Select FacRank, count(*) As NoCourseOfferd, count(distinct Faculty.FacNo) As NoOfFAculty
from Faculty inner join Offering on Faculty.FacNo=Offering.FacNo
Group by FacRank
'''
Q13 =  psql.sqldf(txt)
Q13
```

```python
#For Q14.Q15,Q16 execute this cell first
import sqlite3
  
# Connecting to sqlite
conn = sqlite3.connect('gfs1.db')
  
# Creating a cursor object using 
# the cursor() method
cursor = conn.cursor()
  
# Creating table
table = """CREATE TABLE Student
(StdNo CHAR(11),
StdFirstName VARCHAR(50),
StdLastName VARCHAR(50),
StdCity VARCHAR(50),
StdState CHAR(2),
StdZip CHAR(10),
StdMajor CHAR(6),
StdClass CHAR(6),
StdGPA DECIMAL(3,2),
CONSTRAINT PKStudent PRIMARY KEY (StdNo)  );"""
cursor.execute(table)
```

```python
#For Q14,Q15 and Q16 execute earlier cell first. 
#Q14-Insert a row into the Student table supplying following values for all columns.
#('487-44-1234', 'ELON', 'MUSK','SEATTLE','WA',98121-1111,'FIN','SO',3.9) 
#('457-98-1234', 'BILL', 'GATES','REDMONT','WA',98116-2344,'IS','SR',3.8)
#into
# Student(StdNo,StdFirstName,StdLastName,StdCity,StdState,StdZip ,StdMajor,StdClass,StdGPA) 
cursor.execute(
    '''
    INSERT INTO Student(StdNo,StdFirstName,StdLastName,StdCity,StdState,StdZip ,StdMajor,StdClass,StdGPA)
    VALUES ('487-44-1234', 'ELON', 'MUSK','SEATTLE','WA',98121-1111,'FIN','SO',3.9),
    ('457-98-1234', 'BILL', 'GATES','REDMONT','WA',98116-2344,'IS','SR',3.8)
    ''')
print("Student Table: ")
data = cursor.execute('''SELECT * FROM Student''')
for row in data:
    print(row)
conn.commit()
```
