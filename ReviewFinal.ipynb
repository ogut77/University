```python
#Q5-List  Student LastName and FirstName of the Student with StdGPA greater than average GPA
#Use nested query by calculating average GPA first.
txt='''
Select StdFirstName,	StdLastName	,StdGPA,  (select avg(StdGPA) from Student) AS AvgStdGPA
FROM Student
WHERE StdGPA > (select avg(StdGPA) from Student)
'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#Q5-List  Student LastName and FirstName of the Student with StdGPA greater than average GPA
#Put Averge GPA intoas a column to the Query
#Use nested query by calculating average GPA first.
txt='''
WITH AvgStdGPA AS (
SELECT Avg(StdGPA) as AvgGPA
From Student
)
Select StdFirstName, StdLastName, StdGPA, AvgStdGPA.AvgGPA
From Student, AvgStdGPA
Where StdGPA >(select avg(StdGPA) From Student)
'''
Q5 =  psql.sqldf(txt)
Q5
```

```python
#For Q14.Q15,Q16 execute this cell first
import sqlite3
  
# Connecting to sqlite
conn = sqlite3.connect('gfs1.db')
  
# Creating a cursor object using 
# the cursor() method
cursor = conn.cursor()
  
# Creating table
table = """CREATE TABLE Student
(StdNo CHAR(11),
StdFirstName VARCHAR(50),
StdLastName VARCHAR(50),
StdCity VARCHAR(50),
StdState CHAR(2),
StdZip CHAR(10),
StdMajor CHAR(6),
StdClass CHAR(6),
StdGPA DECIMAL(3,2),
CONSTRAINT PKStudent PRIMARY KEY (StdNo)  );"""
cursor.execute(table)
```
